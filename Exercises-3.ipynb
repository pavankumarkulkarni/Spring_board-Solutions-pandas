```python
titles = pd.read_csv('data/titles.csv', index_col=None)
titles.head()
```

```python
cast = pd.read_csv('data/cast.csv', index_col=None)
cast.head()
```

### Using groupby(), plot the number of films that have been released each decade in the history of cinema.

```python
titles.groupby([titles['year'] // 10*10]).size().sort_index().plot(kind = 'bar')
```

### Use groupby() to plot the number of "Hamlet" films made each decade.

```python
titles[titles['title'] == 'Hamlet'].groupby([titles['year'] // 10 *10]).size().sort_index().plot(kind = 'bar')
```

### How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s?

```python
cast[(cast['n'] == 1) & (cast['year'] // 10*10 == 1950)].groupby(['year','type']).size()
```

### In the 1950s decade taken as a whole, how many total roles were available to actors, and how many to actresses, for each "n" number 1 through 5?

```python
cast[(cast['year'] // 10 *10 == 1950) & (cast['n'] <= 5)].groupby(['n','type']).size()
```

### Use groupby() to determine how many roles are listed for each of the Pink Panther movies.

```python
cast[cast['title'].str.contains('Pink Panther')].groupby(['year','title']).size()
```

### List, in order by year, each of the films in which Frank Oz has played more than 1 role.

```python
cast[cast['name'] == 'Frank Oz'].groupby(['title','year']).filter(lambda x : len(x) > 1).sort_values(by = ['year','title'])
```

### List each of the characters that Frank Oz has portrayed at least twice.

```python
cast[cast['name'] == 'Frank Oz'].groupby(['character']).filter(lambda x : len(x) > 1).sort_values(by=['character'])
```
